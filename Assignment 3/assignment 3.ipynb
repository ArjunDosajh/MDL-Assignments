```python
# Defining all the constants and the gridworld
R = -0.04
discount = 0.95
prob_same_direction = 0.7
prob_perp_direction = 0.15
reward = 1
penalty = -1
cell = {0: [1, 1], 1: [2, 1], 2: [3, 1], 3: [1, 2], 4: [2, 2], 5: [3, 2], 6: [1, 3], 7: [2, 3], 8: [3, 3], 9: [1, 4], 10: [2, 4], 11: [3, 4]}
theta = 0.0001
delta = 100
wall = [2, 2]
reward = 1
penalty = -1
```

```python
# Getting the cell number from a cartesian coordinate
def cartesianToCellNo(cart):
    for x in cell:
        if cell[x] == cart:
            return x
```

```python
# This function calculates sigma(P(J|I,A) * old_V(J)) for all the possible actions
def all_sums(cell_no, old_utils):
    res = []
    # direction == "UP"
    up = cell[cell_no].copy()
    up_sum = 0
    if(up[1] < 4):
        up[1] += 1
    if(up == wall):
        up[1] -= 1
    up_sum += prob_same_direction * old_utils[cartesianToCellNo(up)]
    right = cell[cell_no].copy()
    if(right[0] < 3):
        right[0] += 1
    if(right == wall):
        right[0] -= 1
    up_sum += prob_perp_direction * old_utils[cartesianToCellNo(right)]
    left = cell[cell_no].copy()
    if(left[0] > 1):
        left[0] -= 1
    if(left == wall):
        left[0] += 1
    up_sum += prob_perp_direction * old_utils[cartesianToCellNo(left)]
    res.append(up_sum)

    # direction == "DOWN"
    down = cell[cell_no].copy()
    down_sum = 0
    if(down[1] > 1):
        down[1] -= 1
    if(down == wall):
        down[1] += 1
    down_sum += prob_same_direction * old_utils[cartesianToCellNo(down)]
    right = cell[cell_no].copy()
    if(right[0] < 3):
        right[0] += 1
    if(right == wall):
        right[0] -= 1
    down_sum += prob_perp_direction * old_utils[cartesianToCellNo(right)]
    left = cell[cell_no].copy()
    if(left[0] > 1):
        left[0] -= 1
    if(left == wall):
        left[0] += 1
    down_sum += prob_perp_direction * old_utils[cartesianToCellNo(left)]
    res.append(down_sum)

    # direction == "RIGHT"
    right = cell[cell_no].copy()
    right_sum = 0
    if(right[0] < 3):
        right[0] += 1
    if(right == wall):
        right[0] -= 1
    right_sum += prob_same_direction * old_utils[cartesianToCellNo(right)]
    up = cell[cell_no].copy()
    if(up[1] < 4):
        up[1] += 1
    if(up == wall):
        up[1] -= 1
    right_sum += prob_perp_direction * old_utils[cartesianToCellNo(up)]
    down = cell[cell_no].copy()
    if(down[1] > 1):
        down[1] -= 1
    if(down == wall):
        down[1] += 1
    right_sum += prob_perp_direction * old_utils[cartesianToCellNo(down)]
    res.append(right_sum)

    # direction == "LEFT"
    left = cell[cell_no].copy()
    left_sum = 0
    if(left[0] > 1):
        left[0] -= 1
    if(left == wall):
        left[0] += 1
    left_sum += prob_same_direction * old_utils[cartesianToCellNo(left)]
    up = cell[cell_no].copy()
    if(up[1] < 4):
        up[1] += 1
    if(up == wall):
        up[1] -= 1
    left_sum += prob_perp_direction * old_utils[cartesianToCellNo(up)]
    down = cell[cell_no].copy()
    if(down[1] > 1):
        down[1] -= 1
    if(down == wall):
        down[1] += 1
    left_sum += prob_perp_direction * old_utils[cartesianToCellNo(down)]
    res.append(left_sum)
    
    return res
```

```python
# This function calculates the utility of a cell given the old utilities
def value_function(cell_no, old_utils):
    if cell_no == 10:
        return 1
    elif cell_no == 11:
        return -1
    elif cell_no == 4:
        return 0
    else:
        return R + discount * max(all_sums(cell_no, old_utils))
```

```python
old_V = [0 for _ in range(12)] # old utilities
new_V = [0 for _ in range(12)] # new utilities

# initialising the utilities of the absorbing states and the wall
old_V[10] = 1
old_V[11] = -1
old_V[4] = 0

number_of_iterations = 0

while True:
    number_of_iterations += 1
    ok = True
    for i in range(12):
        new_V[i] = value_function(i, old_V)
        delta = abs(new_V[i] - old_V[i])
        if delta > theta: # Checking the stopping condition
            ok = False
    print("Iteration {}: ".format(number_of_iterations))
    print("{}     {}     {}".format(round(new_V[9], 3), round(new_V[10], 3), round(new_V[11], 3)))
    print("{}     {}     {}".format(round(new_V[6], 3), round(new_V[7], 3), round(new_V[8], 3)))
    print("{}     {}     {}".format(round(new_V[3], 3), round(new_V[4], 3), round(new_V[5], 3)))
    print("{}     {}     {}".format(round(new_V[0], 3), round(new_V[1], 3), round(new_V[2], 3)))
    # for util in new_V:
    #     print(round(util, 3), end='  ')
    print()
    if(ok):
        break
    else:
        old_V = new_V
    new_V = [0 for _ in range(12)]

print("Number of Iterations = ", number_of_iterations)
```

```python
# Finding the optimal policy
optimal_policy = [0 for _ in range(12)]
for i in range(12):
    possible_actions = all_sums(i, new_V)
    optimal_policy[i] = possible_actions.index(max(possible_actions))

for i in range(12):
    if(optimal_policy[i] == 0):
        optimal_policy[i] = "↑"
    elif(optimal_policy[i] == 1):
        optimal_policy[i] = "↓"
    elif(optimal_policy[i] == 2):
        optimal_policy[i] = "→"
    elif(optimal_policy[i] == 3):
        optimal_policy[i] = "←"

print("Optimal Policy is: ")
print("{}    {}    {}".format(optimal_policy[9], "X", "X"))
print("{}    {}    {}".format(optimal_policy[6], optimal_policy[7], optimal_policy[8]))
print("{}    {}    {}".format(optimal_policy[3], "X", optimal_policy[5]))
print("{}    {}    {}".format(optimal_policy[0], optimal_policy[1], optimal_policy[2]))
```
